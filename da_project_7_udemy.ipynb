## **Data Analysis Project 7: Udemy Courses Dataset**
This dataset is primarily centered around the subject or courses offered in Udemy.

Data retrieve from: https://drive.google.com/file/d/1zglnQkX756nfySz9RT8V3p1R6-69RHmS/view

```python
#importing libraries
import pandas as pd
```

```python
#loading dataset directly from Google Drive
url = 'https://drive.google.com/file/d/1zglnQkX756nfySz9RT8V3p1R6-69RHmS/view?usp=sharing'
file_id=url.split('/')[-2]
dwn_url='https://drive.google.com/uc?id=' + file_id
df = pd.read_csv(dwn_url)
df.head()
```

### 7. 'Subject' = 'Graphic Design' and 'Price' < 100

```python
df[df['subject'] == 'Graphic Design'].shape[0]          # condition 1
```

```python
# for condition 2, need to change 'Free' to '0' first
df['price'] = df['price'].replace(['Free'],'0') 
df['price'] = pd.to_numeric(df['price'])
# check data types
df.info()
```

```python
#df[df['price'] < 100]         #condition 2
df[df['price'] < 100].shape[0]
```

```python
# Dataframe based on both condition
df[(df['subject'] == 'Graphic Design') & (df['price'] < 100)].head()
```

```python
# Calculate the result
df[(df['subject'] == 'Graphic Design') & (df['price'] < 100)].shape[0]
```

### 9. Courses where 'year' = 2015

```python
df.info()
```

```python
#change 'published_timestamp' to 'datetime'
df['published_timestamp'] = pd.to_datetime(df['published_timestamp'])
df.info()
```

```python
#insert 'Year'
df.insert(10, 'Year', df.published_timestamp.dt.year)
df.info()
```

```python
df[df['Year'] == 2015].head()
```
