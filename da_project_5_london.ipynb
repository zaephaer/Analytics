## **Data Analysis Project 5: London Housing Dataset**
This dataset is primarily centered around the housing market of London, from 1995 to 2019 in different area. Orignially from Kaggle.

Data retrieve from: https://drive.google.com/file/d/1PenqWMdoFtW53Os_pyalAiIhn0yASY79/view

```python
#importing libraries
import pandas as pd
```

```python
#loading dataset directly from Google Drive
url = 'https://drive.google.com/file/d/1PenqWMdoFtW53Os_pyalAiIhn0yASY79/view?usp=sharing'
file_id=url.split('/')[-2]
dwn_url='https://drive.google.com/uc?id=' + file_id
df = pd.read_csv(dwn_url)
df
```

```python
df.isnull().sum()
```

```python
df.count()
```

### 1. Convert 'Date' to DateTime format

```python
df.info()
```

```python
df.isnull().sum()
```

### 1. Convert 'date' column to 'datetime' format

```python
df['date'] = pd.to_datetime(df['date'])
df.info()
```

### 2. Add column 'year'

```python
#column will add at the end of the dataframe
df['year'] = df['date'].dt.year
df.head()
```

### 3. Add column 'month' as 2nd column

```python
df.insert(1, 'month', df['date'].dt.month)
df.head()
```

### 4. Remove column

```python
#df.drop(['month','year'], axis = 1, inplace = True)
df.drop(['month'], axis = 1, inplace = True)
df.head()
```

### 5. Calculate 'No of Crimes' = 0

```python
#replace NaN value to 0
#df['DataFrame Column'] = df['DataFrame Column'].fillna(0)
df['no_of_crimes'] = df['no_of_crimes'].fillna(0)
df.head()
```

```python
df.info()
```

```python
#len(df[df['no_of_crimes'] == 0])       #other method with same result
df[df['no_of_crimes'] == 0].shape[0]
```

### 6. Highest and Lowest 'Average Price' in 'England'

```python
dfEngland = df[df['area'] == 'england']
dfEngland.head()
```

```python
dfEngland.info()
```

```python
#Lowest average price by year
dfEngland.groupby('year').average_price.min()
```

```python
#Higest average price by year
dfEngland.groupby('year').average_price.max()
```

### 7. Highest and Lowest 'Crime' by 'Area'

```python
hilo_crimes = df.groupby('area').no_of_crimes.max().sort_values(ascending = False)
hilo_crimes[hilo_crimes != 0]
```

### 8. Zero Crimes Area

```python
df.groupby('area').no_of_crimes.min()
```

```python
df[df['no_of_crimes'] == 0].area.value_counts()
```
