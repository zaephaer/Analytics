## **Data Analysis Project 8: Netflix Dataset**
This dataset is primarily information about TV Series and Movies available on Netflix until 2021. This dataset originally collected from Flixable which is a third party search engine, and made available on Kaggle.

Data retrieve from: https://drive.google.com/file/d/1p630pDcSkbQp_WXig8PeEVevZtSo4GQK/view

```python
#importing libraries
import pandas as pd
```

```python
#loading dataset directly from Google Drive
url = 'https://drive.google.com/file/d/1p630pDcSkbQp_WXig8PeEVevZtSo4GQK/view?usp=sharing'
file_id=url.split('/')[-2]
dwn_url='https://drive.google.com/uc?id=' + file_id
df = pd.read_csv(dwn_url)
df
```

```python
df.info()
```

### 1. Drop All Rows that have 'Null' values in any columns

```python
df = df.dropna(how='any',axis=0) 
df.info()
```

### 2. Convert 'Release Date' to 'Datetime' format

```python
df['Release_Date'] = pd.to_datetime(df['Release_Date'])
df.info()
```

### 3. Find 'Duplicate' rows

```python
df[df.duplicated()]
```

```python
#confirmed duplicates rows
df[df['Show_Id'] == 's684']
```

### 4. Remove 'Duplicate' rows

```python
df.drop_duplicates(inplace = True)
df[df.duplicated()]
```
