## **Data Analysis Project 10: Edgeproperty Dataset**
I scrap this dataset from edgeproperty website using webscraper.io. The dataset originally contains over 18k rows. However, a lot of data cleaning need to be done before any analysis intended.

Data retrieve from: https://drive.google.com/file/d/1QdzL07KEQ_nK7f23WCOjRrOTmjCaTx8h/view?usp=sharing

```python
#importing libraries
import pandas as pd
```

```python
#loading dataset directly from Google Drive
url = 'https://drive.google.com/file/d/1QdzL07KEQ_nK7f23WCOjRrOTmjCaTx8h/view?usp=sharing'     
file_id=url.split('/')[-2]
dwn_url='https://drive.google.com/uc?id=' + file_id
df = pd.read_csv(dwn_url)
df.head()
```

## 1. Drop Unwanted columns 

```python
df.drop(['web-scraper-order','web-scraper-start-url','pagination','pagination-href'], axis=1, inplace=True)
df.head()
```

## 2. Remove Unwanted Strings and change to 'Numeric'

```python
#df['colB'] = df['colB'].str.replace(r'\D', '')
df['price'] = df['price'].str.replace(r'\D', '')

#df['col_name'] = pd.to_numeric(df['col_name'])
df['price'] = pd.to_numeric(df['price'])
df.head()
```

```python
#before change to numeric
df.info()
```

## 3. Cleaning 'Number of Bedrooms' column

```python
df['nobeds'].unique()
```

```python
df['nobeds'].nunique()
```

```python
#Replace multiple values with multiple new values for an individual DataFrame column
#df['column name'] = df['column name'].replace(['1st old value','2nd old value',...],['1st new value','2nd new value',...])
df['nobeds'] = df['nobeds'].replace(
              ['4 beds', '3 beds', '3 + 1 beds', '4 + 1 beds', '5 + 1 beds',
              'Studio', '- beds', '2 beds', '1 beds', '5 beds', '6 beds',
              '1 + 1 beds', '4 + 2 beds', '7 beds', '2 + 1 beds', '8 beds',
              '5 + 2 beds', '- + 5 beds', '9 beds', '6 + 1 beds', '11 beds',
              '7 + 1 beds', '12 beds', '13 beds', '10 beds', '3 + 2 beds',
              '10 + 1 beds', '5 + 3 beds', '- + 3 beds', '10 + 3 beds',
              '- + 1 beds', '7 + 2 beds', '6 + 2 beds', '4 + 3 beds', '15 beds',
              '5 + 5 beds', '8 + 1 beds', '14 beds', '15 + 2 beds', '8 + 2 beds',
              '28 beds', '4 + 4 beds'],
              [4,3,4,5,6,1,0,2,1,5,6,2,6,7,3,8,7,5,9,7,11,8,12,13,10,5,11,8,3,13,1,9,8,7,15,10,9,14,17,10,28,8])
```

## 4. Cleaning 'Number of Bathrooms' column

```python
df['nobath'].unique()
```

```python
df['nobath'] = df['nobath'].replace(
              ['3 bath', '1 bath', '2 bath', '4 bath', '6 bath', '- bath',
              '5 bath', '7 bath', '9 bath', '10 bath', '8 bath', '13 bath',
              '11 bath', '15 bath', '12 bath', '14 bath', '28 bath'],
              [3,1,2,4,6,0,5,7,9,10,8,13,11,15,12,14,28])
```

## 5. Cleaning 'Price per sqft' column (New column 'Carpark')

```python
#create new column 'Carpark' and stored if column contain string 'carpark' and 'Null' if not
df.loc[df['sqft_price'].str.contains('carpark'), 'carpark'] = df['sqft_price']

#Remove unwanted string in 'Carpark' and change it to numeric
df['carpark'] = df['carpark'].str.replace(r'\D', '')
df['carpark'] = pd.to_numeric(df['carpark'])
df.head()
```

## 6. Cleaning 'Price per sqft' column (New column 'Price per sqft')

```python
df.loc[df['sqft_price'].str.contains('RM'), 'price_persqft_0'] = df['sqft_price']

#then drop original 'sqft_price' column
df.drop(['sqft_price'], axis=1, inplace=True)
df.head()
```

## 7. Cleaning 'Size' column, move value containing 'RM' to new column

```python
df.loc[df['size'].str.contains('RM'), 'price_persqft_1'] = df['size']

#combine 2 column, using 'combine_first'
df['price_persqft'] = df['price_persqft_0'].combine_first(df['price_persqft_1'])

#then drop column
df.drop(['price_persqft_0','price_persqft_1'], axis=1, inplace=True)

#Remove unwanted string in 'Price Per Sqft' and change it to numeric
df['price_persqft'] = df['price_persqft'].str.replace(r'\D', '')
df['price_persqft'] = pd.to_numeric(df['price_persqft'])

df.tail()
```

## 8. Create new column, 'Size_0', for 'Size'

```python
#if the string contain 'RM' change it to NaN
df.loc[df['size'].str.contains(' sqft'), 'size_0'] = df['size']

#then drop column 'Size'
df.drop(['size'], axis=1, inplace=True)

#Remove unwanted string in 'Size' and change it to numeric, then drop unwanted column
df['size'] = df['size_0'].str.replace(r'\D', '')
df['size'] = pd.to_numeric(df['size'])
df.drop(['size_0'], axis=1, inplace=True)
df.tail()
```
