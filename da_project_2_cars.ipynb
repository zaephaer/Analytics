## **Data Analysis Project 2: Cars Dataset**
This cars dataset contains 15 different columns from cars Make, Model Type, Origin DriveTrain and others.

Data retrieve from: https://drive.google.com/file/d/1MlrQh-fKcL75ni_swveeiRPW_Vn9Q3xw/view

```python
#importing libraries
import pandas as pd
```

```python
#loading dataset directly from Google Drive
url = 'https://drive.google.com/file/d/1MlrQh-fKcL75ni_swveeiRPW_Vn9Q3xw/view?usp=sharing'
file_id=url.split('/')[-2]
dwn_url='https://drive.google.com/uc?id=' + file_id
df = pd.read_csv(dwn_url)
df
```

```python
df.shape
```

### 1. Data Info

```python
df.info()
```

### 2. Counting Null value

```python
df.isnull().sum()
```

### 3. Show rows with Null value

```python
df[pd.isnull(df).any(axis=1)]
```

### 4. Remove rows using Index number

```python
#removing rows
df = df.drop(index = [30,39,161,173])
#checking removed rows
df[pd.isnull(df).any(axis=1)]
```

### 5. Fill Null value in rows 251 and 252 with Mean value

```python
# Must include 'inplace = True' to make change permenant
# using 'round()' to round the number to single digit without decimals
df['Cylinders'].fillna(round(df['Cylinders'].mean(),0), inplace = True)
df.loc[[251,252]]
```

```python
# Confirming no null value left
df.isnull().sum()
```

### 6. Counting value for each 'Make'

```python
df['Make'].value_counts()
```

### 7. Show list of cars where 'Origin' = 'Asia' or 'Europe'

```python
#1  (df['Origin'] = 'Asia')
#2  (df['Origin'] = 'Europe')
#df[(df['Origin'] == 'Asia') | (df['Origin'] == 'Europe')]  # Method 1

# using 'isin'
df[df['Origin'].isin(['Asia','Europe'])]                    # Method 2
```

### 8. Show number of cars where 'Origin' = 'Asia' or 'Europe'

```python
#df[(df['Origin'] == 'Asia') | (df['Origin'] == 'Europe')].shape[0]
df[df['Origin'].isin(['Asia','Europe'])].shape[0]
```

### 9. Removing rows where 'Weight' > 4000

```python
# Dataset shape (before removing 'Weight' > 4000)
df.shape
```

```python
# Calculate rows meet the condition
df[(df['Weight'] > 4000)].shape[0]
```

```python
# Remove rows using '~'
df = df[~(df['Weight'] > 4000)]
```

```python
# Confirm condition removed
df[(df['Weight'] > 4000)]
```

```python
# Recalculate dataset
df.shape
```

### 10. Increase value in column 'MPG City' by 3

```python
# BEFORE increase by 3
df.head()
```

```python
# Increase using 'lambda x: X+3'
df['MPG_City'] = df['MPG_City'].apply(lambda x: x+3)
df.head()
```
